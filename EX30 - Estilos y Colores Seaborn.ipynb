# Estilos y colores

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
propinas = sns.load_dataset('tips')
propinas.head()
```

```python
sns.set_context('notebook',font_scale=1.2)
plt.figure(figsize=(8,4))
sns.set_style('white')
sns.countplot(x='sex',data=propinas,hue='sex')
#Se supone que en el ejercicio 'x='sex'' divide los colores, pero he anadido 'hue='sex'' y ahora si lo hace.
sns.despine(left=True,bottom=True)
```

```python
sns.countplot(x='sex',data=propinas,hue='sex',palette='husl')
```

```python
colores = ["#FF0000", "#ADAAE7"]
sns.countplot(x='sex',data=propinas, palette=colores)
```
