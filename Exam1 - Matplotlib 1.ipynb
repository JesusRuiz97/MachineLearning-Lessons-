```python
import matplotlib.pyplot as plt
def graficar_linea(x, y):
    if len(x) != len(y):
        raise ValueError("Las listas x e y deben tener la misma longitud.")
    x=[1, 2, 3, 4, 5, 6, 7, 8]
    y=[55, 60, 65, 70, 75, 80, 85, 90]
    plt.plot(x,y)
    plt.title("Relaci√≥n entre horas de estudio y calificaciones")
    plt.xlabel("Horas de Estudio")
    plt.ylabel("Calificaci√≥n")
    plt.show()
```
