```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(2,9,10)
x
```

```python
y = x**2
```

```python
y
```

```python
plt.plot(x,y,'lightgreen')
plt.title("Este es mi grafico")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'g')

plt.subplot(2,1,2)
plt.plot(y,x,'blue')
```

```python
#Programacion ortientada a objetos
```

```python
figura = plt.figure()
grafico = figura.add_axes([0.1,0.1,1,1])
grafico.plot(x,y,'green')
```

```python
figura = plt.figure()
grafico1 = figura.add_axes([0.1,0.1,0.9,0.9])
grafico1.plot(x,y,'r')
grafico2 = figura.add_axes([0.15,0.5,0.4,0.4])
grafico2.plot(y,x,'blue')
```
