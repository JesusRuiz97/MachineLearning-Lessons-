import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Clase que representa un sat√©lite
class Satellite:
    def __init__(self, duracion_mision, paneles_sol, carga_util, consumo_diario):
        self.duracion_mision = duracion_mision
        self.paneles_sol = paneles_sol
        self.carga_util = carga_util
        self.consumo_diario = consumo_diario

    def to_dict(self):
        return {
            'duracion_mision_dias': self.duracion_mision,
            'paneles_sol': self.paneles_sol,
            'carga_util': self.carga_util,
            'consumo_diario': self.consumo_diario
        }

# Clase que genera el conjunto de datos sint√©tico
class SatelliteDatasetGenerator:
    def __init__(self, n=300):
        self.n = n
        np.random.seed(42)

    def generate(self):
        duraciones = np.random.randint(100,1001,size=self.n)
        paneles = np.random.uniform(10,100,size=self.n)
        cargas = np.random.uniform(200,2000,size=self.n)
        ruido = np.random.normal(0,1,size=self.n)
        consumos = 5 + (0.01 * duraciones) + (0.002 * cargas) + ruido
        satellites = []
        for i in range(self.n):
            sat = Satellite(
                duracion_mision=duraciones[i],
                paneles_sol=paneles[i],
                carga_util=cargas[i],
                consumo_diario=consumos[i]
            )
            satellites.append(sat)
        return satellites

# Clase que prepara los datos para an√°lisis
class SatelliteDataProcessor:
    def __init__(self, satellites):
        data_list = [sat.to_dict() for sat in satellites]
        self.df = pd.DataFrame(data_list)
        self.df['eficiencia_energia'] = self.df['consumo_diario'] / self.df['paneles_sol']

    def get_dataframe(self):
        return self.df

# Clase para aplicar regresi√≥n lineal
class EnergyConsumptionRegressor:
    def __init__(self):
        self.model = LinearRegression()
        self.coefficients = None
        self.intercept = None

    def fit(self, X, y):
        self.model.fit(X, y)
        self.coefficients = self.model.coef_[0]
        self.intercept = self.model.intercept_
        y_pred = self.model.predict(X)
        return y_pred

    def evaluate(self, y_true, y_pred):
        r_squared = r2_score(y_true, y_pred)
        return r_squared

    def get_coefficients(self):
        return self.coefficients, self.intercept

# Clase que visualiza los datos y el modelo
class SatellitePlotter:
    def __init__(self, df, y_pred):
        self.df = df
        self.y_pred = y_pred
        
    def plot(self):
        plt.figure(figsize=(10, 6))
        scatter = plt.scatter(
            self.df['duracion_mision'],
            self.df['consumo_diario'],
            c=self.df['carga_util'],
            cmap='viridis',
            alpha=0.6,
            label='Satelites (Datos Reales)'
        )
        
        plt.plot(
            self.df['duracion_mision'],
            self.y_pred,
            color='red',
            linewidth=2,
            label='Prediccion (Regresion Lineal)'
        )
        
        cbar = plt.colorbar(scatter)
        cbar.set_label('Carga Util (kg)')
        
        plt.title('Consumo Energetico vs. Duracion de la Mision')
        plt.xlabel('Duracion de la mision (dias)')
        plt.ylabel('Consumo Diario (kWh)')
        plt.legend()
        plt.grid(True)
        
        plt.show()

# Clase principal que orquesta todo
class SatelliteAnalysisExample:
    def run(self):
        # 1. Generar datos
        print("Generando datos sintetiocos de 300 satelites...")
        generator = SatelliteDatasetGenerator(n=300)
        satellites = generator.generate()

        # 2. Procesar datos
        print("Procesando datos y convirtiendo a DataFrame...")
        processor = SatelliteDataProcessor(satellites)
        df = processor.get_dataframe()

        # 3. Regresi√≥n lineal
        print("Aplicando modelo de Regresi√≥n Lineal...")
        X = df[['duracion_mision_dias']]
        y = df['consumo_diario']
        regressor = EnergyConsumptionRegressor()
        y_pred = regressor.fit(X, y)
        
        # 4. Evaluar modelo
        r_squared = regressor.evaluate(y, y_pred)
        coef, intercept = regressor.get_coefficients()

        # 5. Visualizaci√≥n
        print("\n--- Resultados del Modelo ---")
        print(f" Modelo: y = {coef:.4f} * x + {intercept:.4f}")
        print(f" R¬≤ del modelo: {r_squared:.4f}")
        print("-----------------------------\n")
        
        plotter = SatellitePlotter(df, y_pred)
        plotter.plot()

# Ejecutar
if __name__ == "__main__":
    
    example = SatelliteAnalysisExample()
    
    example.run()
